## Q. 4 Understanding histograms and histogram equalization
This sub-project aims to implement **cv2.equalizeHist** Python function from OpenCV library to equalize the histogram.

Download the “magnified-pollen-dark.tif” from Canvas under the data folder. Read the image using the **cv2.imread('image.tif', cv2.IMREAD_GRAYSCALE)**.

Plot the original image (**plt.imshow**) and the result of the equalized histogram.

Plot the histogram distributions before and after equalization. Use **cv2.calcHist([image], [0], None, [256], [0,256])** to first compute the histogram, then plot it using plt.plot


```python
# Read Image
img = cv2.imread('magnified-pollen-dark.tif', cv2.IMREAD_GRAYSCALE)
```

```python
# Perform histogram equalization
img_eq = cv2.equalizeHist(img)
```

```python
# Plotting image
plt.subplot(121)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img_eq, cmap='gray', vmin=0, vmax=255)
plt.title('Histogram Equalized Image')
```

```python
# Plotting histogram
hist = cv2.calcHist([img],[0],None,[256],[0,256])
hist_eq = cv2.calcHist([img_eq],[0],None,[256],[0,256])
plt.subplot(121)
plt.plot(hist)
plt.title('Original Image')
plt.subplot(122)
plt.plot(hist_eq)
plt.title('Histogram Equalized Image')
plt.tight_layout()
```

## Q. 5 Implementing adaptive histograms to reveal a hidden object
This sub-project aims to implement **cv2.createCLAHE().apply(image)** Python function for adaptive histogram equalization.

Download the “hidden-symbols.tif” from Canvas under the data folder. Read the image using the **cv2.imread('image.tif', cv2.IMREAD_GRAYSCALE)**.

Plot the original image and the result of the equalized histogram using histogram equalization and adaptive histogram equalization. No need for additional plots.


```python
# Load image
img2 = cv2.imread('hidden-symbols.tif', cv2.IMREAD_GRAYSCALE)

# Adaptive histogram equalization
img2_adapt = cv2.createCLAHE().apply(img2)

# Plotting
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(img2_adapt, cmap='gray', vmin=0, vmax=255)
plt.title('Histogram Equalized Image')
```
